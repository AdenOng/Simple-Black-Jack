```python
#Global Addresses
#Global Variables with the suits, numbers and the equivalent value
from IPython.display import clear_output
import random
suits = ('Diamonds','Hearts','Clover','Spades')
numbers = ('Ace','Two','Three','Four','Five','Six','Seven','Eight','Nine','Ten','Jack','Queen','King')
blackjack_value = {'Ace': [1,11],'Two': 2,'Three' : 3,'Four' : 4,'Five' : 5,'Six' : 6,'Seven' : 7,'Eight' : 8,'Nine' : 9,'Ten' : 10,'Jack' : 10,'Queen' : 10,'King' : 10}    
```

```python
#creating a class card which has a number and suit
class card:
    def __init__(self,suit,number):
        self.suit = suit
        self.number = number
        self.blackjack_value = blackjack_value[number]
        
    def __str__(self):
        return self.number + ' of ' + self.suit
```

```python
#creating a deck which has a deck of cards
class deck:
    def __init__(self):
        self.all_cards = []
        for suit in suits:
            for number in numbers:
                self.all_cards.append(card(suit,number))
    
    def shuffle(self):
        random.shuffle(self.all_cards)
    
    def draw_1(self):
        return self.all_cards.pop()
```

```python
#function to request bets for the player
def bet(player):
    print('Please place your bets!')
    bet_amount = ''
    while type(bet_amount) != type(2):
        try:
            bet_amount = int(input('Please give me a number '))
            if bet_amount > player.cash:
                print('Sorry the amount you betted is greater than what you have')
                bet_amount = ''
        except:
            print('Sorry please give me a number!')
    return bet_amount
```

```python
#function to exit the game
def replay():
    gameon = ''
    while gameon not in ['Y','N']:
        gameon = input('Do you want to carry on? Please choose Y/N.')
        if gameon not in ['Y','N']:
            clear_output()
            print('Sorry please choose Y/N. Do you want to carry on? Please choose Y/N.')
    if gameon == 'Y':
        return True
    if gameon == 'N':
        return False
```

```python
#function to make decision
def decision(player):
    choice = ""
    while choice.upper() not in ['STAY','HIT','SPLIT']:
        choice = input('Choose the following STAY,HIT or SPLIT.')
        if player.sumation_card() <= 15:
            if choice.upper() in ['STAY']:
                print('Sorry you cannot choose to stay as the value of your card is under 15! Please input a different option!')
                choice = ""
        if player.all_cards[0].blackjack_value != player.all_cards[1].blackjack_value:
            if choice.upper() in ['SPLIT']:
                print('Sorry both are not of the same value you cannopt split')
    return choice.upper()
        
    
```
